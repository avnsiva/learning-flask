```python
def annual_volatility(df):
    df['logreturns'] = np.log(df['Close'] / df['Close'].shift(1))
    df['AnnualVolatility']=0
    i = 10
    for eachrows in df['logreturns'][10:]:
        subret = df['logreturns'][(i-10):i]
        df['AnnualVolatility'][i] = (np.sqrt(252*subret.var())*100)
        i = i+1
    return df
```

```python
niftydf = annual_volatility(niftydf)

print niftydf
```

```python
ix = 2
df = niftydf
print df.count
#df['Crossing']= 'null'
niftydf['EMA20_5_LongSignal']= 'null'
niftydf['EMA20_5_ShortSignal']= 'null'
niftydf['EMA5_Close_ShortSignal']= 'null'
niftydf['EMA5_Close_LongSignal']= 'null'
niftydf['ATR14_Close_ShortSignal']= 'null'
niftydf['ATR14_Close_LongSignal']= 'null'

for eachrow in niftydf['logreturns'][2:]:
    previous_20=niftydf['EMA_20'][ix-1]
    previous_5=niftydf['EMA_5'][ix-1]
    previous_close = niftydf['Close'][ix-1]
    previous_atr = niftydf['ATR_14'][ix-1]
    #crossing = (df['EMA_5'][ix] <= df['EMA_20'][ix])
    EMA20_5_ShortSignal = ((niftydf['EMA_5'][ix] <= niftydf['EMA_20'][ix]) & (previous_5 >= previous_20))
    EMA20_5_LongSignal = ((niftydf['EMA_5'][ix] >= niftydf['EMA_20'][ix]) & (previous_5 <= previous_20))
    EMA5_Close_ShortSignal =  ((niftydf['Close'][ix] <= niftydf['EMA_5'][ix]) & (previous_close >= previous_5))
    EMA5_Close_LongSignal =  ((niftydf['Close'][ix] >= niftydf['EMA_5'][ix]) & (previous_close <= previous_5))
    ATR14_Close_ShortSignal =  ((niftydf['Close'][ix] <= niftydf['ATR_14'][ix]) & (previous_close >= previous_atr))
    ATR14_Close_LongSignal =  ((niftydf['Close'][ix] >= niftydf['ATR_14'][ix]) & (previous_close <= previous_atr))
    #crossing = (((df['EMA_5'][ix] <= df['EMA_20'][ix]) & (previous_5 >= previous_20)) | ((df['EMA_5'][ix] >= df['EMA_20'][ix]) & (previous_5 <= previous_20)))
    if EMA20_5_LongSignal == True:
        niftydf['EMA20_5_LongSignal'][ix] = df['Close'][ix]
    if EMA20_5_ShortSignal == True:
        niftydf['EMA20_5_ShortSignal'][ix] = df['Close'][ix]
    if EMA5_Close_LongSignal == True:
        niftydf['EMA5_Close_LongSignal'][ix] = df['Close'][ix]
    if EMA5_Close_ShortSignal == True:
        niftydf['EMA5_Close_ShortSignal'][ix] = df['Close'][ix]
    if ATR14_Close_LongSignal == True:
        niftydf['ATR14_Close_LongSignal'][ix] = df['Close'][ix]
    if ATR14_Close_ShortSignal == True:
        niftydf['ATR14_Close_ShortSignal'][ix] = df['Close'][ix]
    ix = ix + 1
```
